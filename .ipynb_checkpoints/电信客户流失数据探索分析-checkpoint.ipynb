```python
import numpy as np
import pandas as pd 
```

```python
telecom = pd.read_csv('WA_Fn-UseC_-Telco-Customer-Churn.csv')
telecom.head()
```

```python
telecom.shape
telecom.describe()
telecom.info()
telecom.isnull().sum()#不存在空值
```

```python
# customerID ：用户ID。
# gender：性别。（Female & Male）
# SeniorCitizen ：老年人 （1表示是，0表示不是）
# Partner ：是否有配偶 （Yes or No）
# Dependents ：是否经济独立 （Yes or No）
# tenure ：客户的职位（0-72，共73个职位）
# PhoneService ：是否开通电话服务业务 （Yes or No）
# MultipleLines：是否开通了多线业务（Yes 、No or No phoneservice 三种）
# InternetService：是否开通互联网服务 （No, DSL数字网络，fiber optic光纤网络 三种）
# OnlineSecurity：是否开通网络安全服务（Yes，No，No internetserive 三种）
# OnlineBackup：是否开通在线备份业务（Yes，No，No internetserive 三种）
# DeviceProtection：是否开通了设备保护业务（Yes，No，No internetserive 三种）
# TechSupport：是否开通了技术支持服务（Yes，No，No internetserive 三种）
# StreamingTV：是否开通网络电视（Yes，No，No internetserive 三种）
# StreamingMovies：是否开通网络电影（Yes，No，No internetserive 三种）
# Contract：签订合同方式 （按月，一年，两年）
# PaperlessBilling：是否开通电子账单（Yes or No）
# PaymentMethod：付款方式（bank transfer，credit card，electronic check，mailed check）
# MonthlyCharges：月费用
# TotalCharges：总费用
# Churn：该用户是否流失（Yes or No）
```

```python
#可视化
import matplotlib.pyplot as plt
import seaborn 
from matplotlib import font_manager
my_font = font_manager.FontProperties(fname='/System/Library/Fonts/PingFang.ttc',size=10)
```

```python
#饼图 
labels = telecom.Churn.value_counts().index
churn_value = telecom.Churn.value_counts()

# rcParams('figure.figsize') = 6,6

plt.axes(aspect='equal')

plt.pie(churn_value,
        explode=(0.1,0),
        labels = labels,
        colors=['red','orange'],
        shadow=True,
        autopct="%1.1f%%",
        radius=1.0,
        textprops= {'fontsize':12,'color':'white'})

plt.title('Proportions of Customer Churn')
plt.show()

#26.8%的用户流失
```

```python
fig,axes = plt.subplots(2,2,figsize=(10,10))

plt.subplot(2,2,1)
gender=seaborn.countplot(x="gender",hue="Churn",data=telecom,palette="Pastel2")
plt.xlabel("gender")
plt.title("Churn by Gender")

plt.subplot(2,2,2)
gender=seaborn.countplot(x="SeniorCitizen",hue="Churn",data=telecom,palette="Pastel2")
plt.xlabel("SeniorCitizen")
plt.title("Churn by SeniorCitizen")

plt.subplot(2,2,3)
gender=seaborn.countplot(x="Partner",hue="Churn",data=telecom,palette="Pastel2")
plt.xlabel("Partner")
plt.title("Churn by Partner")

plt.subplot(2,2,4)
gender=seaborn.countplot(x="Dependents",hue="Churn",data=telecom,palette="Pastel2")
plt.xlabel("Dependents")
plt.title("Churn by Dependents")

# 可以看出，男性与女性用户之间的流失情况基本没有差异，而在老年用户中流失占比明显比非老年用户更高，非老年用户流失率23.65%
#而老年用户流失率占比达到41.68%
# 在所有数据中未婚与已婚人数基本持平，但未婚中流失人数比已婚中的流失人数高出了快一倍，从经济独立情况来看，
# 经济未独立的用户流失率要远远高于经济独立的用户。

```

```python
#非老龄客户流失比例
NoSenior = telecom[telecom.SeniorCitizen == 0]
NoSenior_Churn = NoSenior[NoSenior.Churn == 1].shape[0]/NoSenior.shape[0]
print("%.2f%%" % (NoSenior_Churn * 100))
#23.65%
```

```python
#老龄客户流失比例
Senior = telecom[telecom.SeniorCitizen == 1]
Senior_Churn = Senior[Senior.Churn == 1].shape[0]/Senior.shape[0]
print("%.2f%%" % (Senior_Churn * 100))
#41.68%
```

```python
#独立客户流失率
Dependents = telecom[telecom.Dependents == 'Yes']
Dependents_Churn = Dependents[Dependents.Churn == 1].shape[0]/Dependents.shape[0]
print("%.2f%%" % (Dependents_Churn * 100))
```

```python
#独立客户流失率
No_Dependents = telecom[telecom.Dependents == 'No']
No_Dependents_Churn = No_Dependents[No_Dependents.Churn == 1].shape[0]/No_Dependents.shape[0]
print("%.2f%%" % (No_Dependents_Churn * 100))
```

```python
#考虑tenure
#范围1-72 
# telecom.tenure.value_counts()
#group by tenure
group_tenure  = telecom.groupby(by = 'tenure').count().Churn.sort_values()
group_tenure

#group by tenure and Churn == 1
Yes_tenure = telecom[telecom.Churn == 1]
Yes_group_tenure  = Yes_tenure.groupby(by = 'tenure').count().Churn.sort_values()
Yes_group_tenure

concat_tenure = pd.concat([group_tenure,Yes_group_tenure],axis=1)
concat_tenure['percent'] = concat_tenure.iloc[:,1]/concat_tenure.iloc[:,0]

#
```

```python
# MultipleLines：是否开通了多线业务（Yes 、No or No phoneservice 三种）
# InternetService：是否开通互联网服务 （No, DSL数字网络，fiber optic光纤网络 三种）
# OnlineSecurity：是否开通网络安全服务（Yes，No，No internetserive 三种）
# OnlineBackup：是否开通在线备份业务（Yes，No，No internetserive 三种）
# DeviceProtection：是否开通了设备保护业务（Yes，No，No internetserive 三种）
# TechSupport：是否开通了技术支持服务（Yes，No，No internetserive 三种）
# StreamingTV：是否开通网络电视（Yes，No，No internetserive 三种）
# StreamingMovies：是否开通网络电影（Yes，No，No internetserive 三种）
# Contract：签订合同方式 （按月，一年，两年）
# PaperlessBilling：是否开通电子账单（Yes or No）
# PaymentMethod：付款方式（bank transfer，credit card，electronic check，mailed check）
covariables=["OnlineSecurity", "OnlineBackup", "DeviceProtection", "TechSupport", "StreamingTV", "StreamingMovies"]
fig,axes=plt.subplots(nrows=2,ncols=3,figsize=(16,10))
for i, item in enumerate(covariables):
    plt.subplot(2,3,(i+1))
    ax=seaborn.countplot(x=item,hue="Churn",data=telecom,palette="Pastel2",order=["Yes","No","No internet service"])
    plt.xlabel(str(item))
    plt.title("Churn by "+ str(item))
    i=i+1
plt.show()
```

```python
#分析上图
#for online security

Yes_OnlineSecurity = telecom[telecom.OnlineSecurity == 'Yes'].shape[0]
#telecom[telecom[telecom.OnlineSecurity == 'Yes'].Churn == 1].shape
Yes_OnlineSecurity_Churn = telecom[telecom.OnlineSecurity == 'Yes'][telecom[telecom.OnlineSecurity == 'Yes'].Churn == 1].shape[0]
Yes_OnlineSecurity_Churn_Percent = Yes_OnlineSecurity_Churn/Yes_OnlineSecurity
print("办理网络安全服务的客户流失率为：%.2f%%" % (Yes_OnlineSecurity_Churn_Percent*100))
#14.64% 办理网络安全服务的客户流失率为14.64% 

No_OnlineSecurity = telecom[telecom.OnlineSecurity == 'No'].shape[0]
#telecom[telecom[telecom.OnlineSecurity == 'Yes'].Churn == 1].shape
No_OnlineSecurity_Churn = telecom[telecom.OnlineSecurity == 'No'][telecom[telecom.OnlineSecurity == 'No'].Churn == 1].shape[0]
No_OnlineSecurity_Churn_Percent = No_OnlineSecurity_Churn/No_OnlineSecurity
print("未办理网络安全服务的客户流失率：%.2f%%" % (No_OnlineSecurity_Churn_Percent*100))
#41.78% 未办理网络安全服务的客户流失率为14.64% 

NoService_OnlineSecurity = telecom[telecom.OnlineSecurity == 'No internet service'].shape[0]
#telecom[telecom[telecom.OnlineSecurity == 'Yes'].Churn == 1].shape
NoService_OnlineSecurity_Churn = telecom[telecom.OnlineSecurity == 'No internet service'][telecom[telecom.OnlineSecurity == 'No internet service'].Churn == 1].shape[0]
NoService_OnlineSecurity_Churn_Percent = NoService_OnlineSecurity_Churn/NoService_OnlineSecurity
print("无网络服务的客户流失率为：%.2f%%" % (NoService_OnlineSecurity_Churn_Percent*100))
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(20,10))

plt.subplot(1,2,1)
seaborn.barplot(x="Contract",y="Churn", data=telecom, palette="plasma_r", ci=95,errcolor='yellow', errwidth=2, alpha=0.3, order= ['Month-to-month', 'One year', 'Two year'])
plt.title("Churn by Contract type")

plt.subplot(1,2,2)
seaborn.countplot(x="Contract",hue="Churn", data=telecom, palette="plasma_r",alpha=0.3,order= ['Month-to-month', 'One year', 'Two year'])
plt.title("Churn by Contract type")

# 由图上可以看出，签订合同方式对客户流失率影响为：按月签订 > 按一年签订 > 按两年签订，这可能表明，设定长期合同对留住现有客户更有效。
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(20,10))
plt.subplot(1,2,1)
seaborn.barplot(x="PaymentMethod",y="Churn", data=telecom, palette="Pastel1", order= ['Bank transfer (automatic)', 'Credit card (automatic)', 'Electronic check','Mailed check'])
plt.title("Churn by PaymentMethod type")
plt.subplot(1,2,2)
seaborn.countplot(x="PaymentMethod",hue="Churn", data=telecom, palette="plasma_r",alpha=0.3)
plt.title("Churn by Contract type")

#由下图可以看出，四种支付方式中，使用Electronic check支付方式的客户流失率最高，其他三种支付方式基本持平，因此可以推断电子账单在设计上影响用户体验。
```
